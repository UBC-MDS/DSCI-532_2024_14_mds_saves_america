```python
import pandas as pd
```

```python
# Loading the raw data
path = 'data/raw/201808-CAH_PulseOfTheNation_Raw.csv'
data = pd.read_csv(path)

data.head()
```

### Raw Data EDA 

```python
# Checking Column Names and Types
print("Column Names and Types:\n", data.dtypes)
```

```python
# Row and Column Counts
num_rows, num_columns = data.shape
print("\nNumber of Rows:", num_rows)
print("Number of Columns:", num_columns)
```

```python
# Calculate the total number of missing values

# Calculate the total number of cells
total_cells = data.size

# Calculate the total number of missing values
total_missing_values = data.isna().sum().sum()

# Calculate the percentage of missing values
missing_percentage = round((total_missing_values / total_cells) * 100)

# Print the percentage of missing values
print("Percentage of missing values:", missing_percentage)
```

```python
# Summary Statistics
column_properties = data.describe(include='all')
print("\nSummary Statistics for Each Column:\n", column_properties)
```

### Data Preprocessing

```python
# Drop irrelevant/redundant columns from the dataset
columns_to_drop = ['Q3OTH1', 'Q4OTH', 'Q2: What is your age?', 
                   'Q16: If Q15 = Yes... And do you think America will survive the next thousand years?', 
                   'Q14: If Q13 = Yes... And do you think humanity will survive the next thousand years?']

# Dropping the columns
data_dropped = data.drop(columns=columns_to_drop, errors='ignore')

# Displaying the first few rows of the modified dataset
data_dropped_head = data_dropped.head()

data_dropped_head
```

```python
new_column_names = ['gender', 'age_range', 'race', 'higher_education', 'political_party', 'ideology', 
                    'trump_approval', 'monetary_anxiety', 'fairness_voting', 'preferential_voting', 
                    'woman_president_25', 'universal_healthcare_25', 'humanity_survival_100', 
                     'america_survival_100', 'trump_2020', 'trump_nuclear', 'america_vs_russia_50',
                    'america_vs_china_50', 'better_world']

# Updating the column names
data_dropped.columns = new_column_names
data_renamed = data_dropped

# Display the first few rows of the updated dataframe
data_renamed_head = data_renamed.head()

data_renamed_head
```

```python
# Create a new column called political ideaology score to reflect political ideaology standing from a numerical scale perspective
# Map the 'ideology' values to scores
ideology_to_score = {
    'Conservative': -1,  # Assuming a lower score for conservative
    'Liberal': 1,        # Assuming a higher score for liberal
    'Moderate': 0,       # Neutral score for moderate
    'DK/REF': 'DK/REF'   # Keeping DK/REF as is
}

# Apply the mapping to create the new column
data_renamed['political_spectrum_score'] = data_renamed['ideology'].map(ideology_to_score)

# Display the first few rows of the updated dataframe
print(data_renamed.head())
```

```python
#add a time period column 

data_renamed['time_answered'] = pd.to_datetime('August 2018').strftime('%Y-%m')
data_cleaned = data_renamed

# Displaying the first few rows to confirm the addition of the new column in datetime format
data_cleaned_head = data_cleaned.head()
data_cleaned

```

```python
# Calculate the total number of missing values

# Calculate the total number of cells
total_cells = data_cleaned.size

# Calculate the total number of missing values
total_missing_values = data_cleaned.isna().sum().sum()

# Calculate the percentage of missing values
missing_percentage = round((total_missing_values / total_cells) * 100)

# Print the percentage of missing values
print("Percentage of missing values:", missing_percentage)
```

```python
# Row and Column Counts
num_rows, num_columns = data_cleaned.shape
print("\nNumber of Rows:", num_rows)
print("Number of Columns:", num_columns)
```

```python
# output cleaned data
output_file_path = 'data/processed/data_cleaned.csv' 

# Save the dataframe to a CSV file without the index
data_dropped.to_csv(output_file_path)

print("Data saved to:", output_file_path)
```
